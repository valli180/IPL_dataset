```python
first_batting_wins = ipl[ipl["win_by_runs"]!=0]
first_batting_wins
```

```python
total_in_city = ipl["city"].value_counts()
total_in_city_df = total_in_city.to_frame().reset_index()
total_in_city_df.columns = ["city", "count"]
total_in_city_df
```

```python
wins_in_city = first_batting_wins.groupby("city").id.count()
wins_in_city_df = wins_in_city.to_frame().reset_index()
wins_in_city_df.columns = ["city", "count"]
print(type(wins_in_city_df))
```

```python
stats_by_city = total_in_city_df.merge(wins_in_city_df, right_on = "city", left_on = "city")
stats_by_city.columns = ["city", "total_played", "total_won"]
print(type(stats_by_city))
```

```python
# Luck with Toss

luck_with_toss = ipl["toss_winner"].value_counts()
luck_with_toss_df = luck_with_toss.to_frame().reset_index()
luck_with_toss_df.columns = ["Team", "count"]
print(type(luck_with_toss_df))
```

```python
luck_with_toss_plt = alt.Chart(luck_with_toss).mark_bar().encode(
                                        x = alt.X("Team", sort = "-y")
                                
                                                 )
         
```
